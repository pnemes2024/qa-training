```python
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium import plugins
from tqdm import tqdm
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

### Dataset

```python
war_data0 = pd.read_csv("/Users/pneme/Documents/Kaggle/ukr-civharm-2023-03-25.csv")
print(war_data0.columns.tolist())
display(war_data0[0:5])

#/Users/pneme/Documents/Kaggle
```

```python
war_data0['weapons1']=war_data0['associations'].apply(lambda x:x.split(',')[-1].split('=')[-1])
war_data0['weapons2']=war_data0['associations'].apply(lambda x:x.split(',')[-1].split('=')[-2])
war_data1=war_data0[['date', 'latitude', 'longitude', 'weapons1','weapons2']]
war_data2=war_data1[war_data1['weapons1']!='Unknown'][war_data1['weapons2']=='Weapon System']
display(war_data2)
display(war_data2['weapons1'].value_counts())
print()
print(war_data2['weapons1'].value_counts().keys().tolist())
weapons=war_data2['weapons1'].value_counts().keys().tolist()
```

## Location Maps and Weapons Used

### Cluster Munitions

```python
wep = weapons[0]
print(wep)
war_data3=war_data2[war_data2['weapons1']==wep]
war_data3m=war_data3[['latitude','longitude']].dropna()
eq_map1 = folium.Map(location=[50,36],tiles='Stamen Terrain',zoom_start=5.0,min_zoom=2.0)
eq_map1.add_child(plugins.HeatMap(war_data3m))
eq_map1
```
